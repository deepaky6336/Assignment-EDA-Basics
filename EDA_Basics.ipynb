## **Missing Values Handle**



```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df = df.dropna()
```

## ***Q1. Average, Median and Standard Deviation of House Prices***

---



```python
print("Average Price =", df['Price'].mean())
print("Median Price =", df['Price'].median())
print("Standard Deviation =", df['Price'].std())
```

## ***Interpretation***

### *The average house price is higher than the median price, which indicates that a few very expensive houses are increasing the overall average. This suggests the presence of high-value properties in the dataset.*



## ***Q2. Which number of bedrooms is most common?***

```python
df['number of bedrooms'].mode()
```

## ***Interpretation***

### *The most common house type has 3 bedrooms. This shows that 3-bedroom houses are the most preferred or most frequently available properties in the dataset.*

## ***Q3. Check if the price data is skewed***

```python
df['Price'].skew()
```

## ***Interpretation***

### *The price distribution is positively skewed (right-skewed). Most houses have relatively lower prices, while a small number of houses have exceptionally high prices.*

## ***Q4. Average price for each number of bedrooms***

```python
df.groupby('number of bedrooms')['Price'].mean()
```

## ***Interpretation***

### *The average house price increases as the number of bedrooms increases. Larger houses generally have higher market values.*

## ***Q5. Relationship between Living Area and Price***

### ***(Use: Area of the house excluding basement)***

```python
df['Area of the house(excluding basement)'].corr(df['Price'])
```

## ***Interpretation***

### *There is a moderate positive relationship between living area and house price. Houses with larger living areas tend to have higher prices.*

## ***Q6. Identify anomalies where houses have high prices but low areas***

```python
anomalies = df[
    (df['Price'] > df['Price'].quantile(0.90)) &
    (df['Area of the house(excluding basement)'] <
     df['Area of the house(excluding basement)'].quantile(0.25))
]
print(len(anomalies))
```

## ***Interpretation***

### *A few houses have unusually high prices despite having smaller living areas. These properties may be located in premium locations or may include special features that increase their value.*

## ***Q7. Compare average price based on number of floors and waterfront***

```python
df.groupby(['number of floors','waterfront present'])['Price'].mean()
```

## ***Interpretation***

### *Houses with waterfront views have significantly higher average prices than houses without waterfront views. Waterfront location has a strong impact on property value.*

## ***Q8. Minimum and Maximum House Price***

```python
print(df['Price'].min())
print(df['Price'].max())
```

## ***Interpretation***

### *The large difference between the minimum and maximum house prices indicates a wide variation in property values within the dataset.*

## ***Q9. Which location (zipcode/postal code) has the highest average price?***

```python
df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head(1)
```

## ***Interpretation***

### *The postal code with the highest average house price represents the most expensive area in the dataset. Properties in this location are generally valued much higher than in other areas.*
